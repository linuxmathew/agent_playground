```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class AgentState(TypedDict):
    name: str
    result:str
```

```python
def hello_node(state: AgentState)->AgentState:
    """This function greets the user"""
    state['result'] = f"Hello, {state['name']}. How are you doing today?"
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node('hello', hello_node)
graph.set_entry_point('hello')
graph.set_finish_point('hello')
# graph.add_edge(START, 'hello')
# graph.add_edge('hello', END)

builder = graph.compile()
```

```python
from IPython.display import Image, display

display(Image(builder.get_graph().draw_mermaid_png()))
```

```python
output = builder.invoke({ 'name': "Temitayo"})
print(output['result'])
```
